```python
from groq import Groq
from google.colab import userdata


api_key = userdata.get("TASK1")
client = Groq(api_key=api_key)


system_message = {
    "role": "system",
    "content": (
        "You are a travel enthusiast and will help people plan vacations. "
        "You will:\n"
        "1. Suggest destinations based on the user's request.\n"
        "2. Provide detailed itineraries with daily activities and points of interest.\n"
        "3. Also be considerate of the user's budget or tell them how much should they atleast have for a certain destination. \n"
        "4. When helping them plan a vacation to certain destination teach them some words of the native language and crack some jokes. \n"
        "5. If a user is struggling to convey their message to the locals at the destination help him/her with it. \n                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                             "
        "When you're ready, greet the user and ask how you can help."
    )
}


model_name = "llama-3.1-8b-instant"


initial = client.chat.completions.create(
    model=model_name,
    messages=[system_message]
)
print("Bot:", initial.choices[0].message.content)


while True:
    user_input = input("You: ")
    if user_input.strip().lower() == "exit":
        print("Bot: Goodbye! Have a wonderful trip planning ahead.")
        break

    response = client.chat.completions.create(
        model=model_name,
        messages=[system_message, {"role": "user", "content": user_input}],
        temperature=0.7
    )

    print("Bot:", response.choices[0].message.content)
```
